```python
# ðŸ“Œ Enhanced EDA Pipeline
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Load dataset
col_names = [
    "age", "sex", "cp", "trestbps", "chol", "fbs",
    "restecg", "thalach", "exang", "oldpeak", "slope",
    "ca", "thal", "target"
]

df = pd.read_csv(
    "/kaggle/input/datata/processed.cleveland.data",
    names=col_names
)

# Replace "?" with NaN and convert to numeric where needed
df = df.replace("?", pd.NA)
df = df.apply(pd.to_numeric, errors="ignore")

print("âœ… Data Loaded")
print(df.head(), "\n")

# 2. Basic info
print("Shape:", df.shape)
print("\nMissing values:\n", df.isnull().sum())
print("\nSummary statistics:\n", df.describe(include="all"))

# 3. Univariate Analysis
plt.figure(figsize=(6,4))
sns.histplot(df["age"].dropna(), bins=20, kde=True)
plt.title("Age Distribution")
plt.tight_layout()
plt.show()

plt.figure(figsize=(6,4))
sns.countplot(x="sex", data=df)
plt.title("Sex Distribution (0=Female, 1=Male)")
plt.tight_layout()
plt.show()

plt.figure(figsize=(6,4))
sns.countplot(x="cp", data=df)
plt.title("Chest Pain Types")
plt.tight_layout()
plt.show()

# 4. Bivariate Analysis
plt.figure(figsize=(6,4))
sns.scatterplot(x="age", y="chol", hue="target", data=df, palette="Set1")
plt.title("Cholesterol vs Age (by Target)")
plt.tight_layout()
plt.show()

plt.figure(figsize=(6,4))
sns.boxplot(x="target", y="thalach", data=df, palette="Set2")
plt.title("Max Heart Rate by Target")
plt.tight_layout()
plt.show()

# 5. Correlation Heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()

# 6. Multivariate
sns.pairplot(df[["age","chol","thalach","target"]].dropna(), hue="target", diag_kind="kde", palette="husl")
plt.suptitle("Pairplot of Key Features", y=1.02)
plt.show()
```
